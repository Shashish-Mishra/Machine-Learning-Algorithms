```python
from sklearn.model_selection import train_test_split
```

```python
x=df[['sepal length (cm)',
 'sepal width (cm)',
 'petal length (cm)',
 'petal width (cm)']]
```

```python
x
```

```python
y=df['target']
```

```python
y
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
```

```python
len(x_train)
```

```python
len(x_test)
```

```python
from sklearn.svm import SVC
```

```python
model=SVC()
```

```python
model.fit(x_train,y_train)
```

```python
model.score(x_test,y_test)
```

```python
model.predict([[4.8,3.0,1.5,0.3]])
```

**tune parameters**

**1. Regularization(C)**

```python
model_c=SVC(C=1)
model_c.fit(x_train,y_train)
model_c.score(x_test,y_test)
```

```python
model_c=SVC(C=10)
model_c.fit(x_train,y_train)
model_c.score(x_test,y_test)
```

**gamma**

```python
model_g=SVC(gamma=10)
model_g.fit(x_train,y_train)
model.score(x_test,y_test)
```

```python
model_g=SVC(gamma=1,C=1)
model_g.fit(x_train,y_train)
model.score(x_test,y_test)
```

**Kernel**

```python
model_liner_kernel=SVC(kernel='linear')
model_liner_kernel.fit(x_train,y_train)
```

```python
model_liner_kernel.score(x_test,y_test)
```
